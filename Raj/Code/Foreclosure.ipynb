```python
import pandas as pd
import numpy as np
import os
import csv
```

```python
# Read the CSV file
csv_path = "../Source/ForeclosureData/Registered_Foreclosure_Properties_2016-2018_Combined.csv"
forcls_df = pd.read_csv(csv_path)

forcls_df = forcls_df[np.isfinite(forcls_df['PropertyZip'])]
```

```python
# Remove NaN rows
forcls_df.dropna(axis=1,how='any')

# Drop Columns
forcls_df.drop(columns=['Lender'],inplace=True)
```

```python
# Chagne Data Type
forcls_df['PropertyZip'] = forcls_df['PropertyZip'].astype(int)
forcls_df['RegisteredDate'] = forcls_df['RegisteredDate'].astype('datetime64')
```

```python
# Converting from date to year
forcls_df['RegisteredDate']=forcls_df['RegisteredDate'].apply(lambda yr:yr.year)
```

```python
# Grouping data by PropertyZip, RegisteredDate and CouncilDistrict
forcls_df.groupby(['CouncilDistrict','PropertyZip','RegisteredDate']).count()
```
