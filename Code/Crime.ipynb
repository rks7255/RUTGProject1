```python
import pandas as pd
import numpy as np
import os
import csv
```

```python
# Read the CSV file
csv_path = "../Source/CrimeData/crimedata2016to2018.csv"
crmcls_df = pd.read_csv(csv_path)

crmcls_df = crmcls_df[np.isfinite(crmcls_df['Zip Codes'])]
```

```python
# Remove NaN rows
crmcls_df.dropna(axis=1,how='any')

# Drop Columns
crmcls_df.drop(columns=['Date Reported','Reporting District','Cross Street','Location ','MO Codes',
                        'Status Code','Status Description','Crime Code 1','Crime Code 2','Crime Code 3',
                        'Crime Code 4'],inplace=True)
```

```python
# Change Data Type
crmcls_df['Zip Codes'] = crmcls_df['Zip Codes'].astype(int)
crmcls_df['Date Occurred'] = pd.to_datetime(crmcls_df['Date Occurred'])
```

```python
# Converting from date to year
crmcls_df['Date Occurred'] = crmcls_df['Date Occurred'].apply(lambda yr:yr.year)
```

```python
# Filter data down to 2016 - 2018
crmcls_df = crmcls_df[crmcls_df["Date Occurred"] > 2015]
```

```python
# Grouping data by Zip Code and Date Occurred Year
crmcls_df.groupby(['Zip Codes','Date Occurred']).count()
```
