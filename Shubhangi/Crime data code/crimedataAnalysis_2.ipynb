```python
import pandas as pd
import numpy as np
```

```python
crimedata_df = pd.read_csv("../Crimedata Output/Crime_Data_LA_Use.csv")
crimedata_df['Area Name'].unique()
```

```python
#grab foreclosure data from 2016
crime16=crimedata_df[crimedata_df['Date Occurred Year']==2016]

#grab total foreclosure for 2016
tot_crime16=crime16["Area Name"].count()


#grab total crime for each area in 2016
seventyseven16crime=crime16["Area Name"].value_counts()["77th Street"]
central16=crime16["Area Name"].value_counts()["Central"]
dev16=crime16["Area Name"].value_counts()["Devonshire"]
fhill16=crime16["Area Name"].value_counts()["Foothill"]
harb16=crime16["Area Name"].value_counts()["Harbor"]
hbeck16=crime16["Area Name"].value_counts()["Hollenbeck"]
hwood16=crime16["Area Name"].value_counts()["Hollywood"]
mission16=crime16["Area Name"].value_counts()["Mission"]
nhwood16=crime16["Area Name"].value_counts()["N Hollywood"]
newton16=crime16["Area Name"].value_counts()["Newton"]
northeast16=crime16["Area Name"].value_counts()["Northeast"]
pacific16=crime16["Area Name"].value_counts()["Pacific"]
rampart16=crime16["Area Name"].value_counts()["Rampart"]
seast16=crime16["Area Name"].value_counts()["Southeast"]
swest16=crime16["Area Name"].value_counts()["Southwest"]
topanga16=crime16["Area Name"].value_counts()["Topanga"]
vnuys16=crime16["Area Name"].value_counts()["Van Nuys"]
wla16=crime16["Area Name"].value_counts()["West LA"]
wvalley16=crime16["Area Name"].value_counts()["West Valley"]
wilshire16=crime16["Area Name"].value_counts()["Wilshire"]


# Percentage of crime per area
per_seventyseven16crime = seventyseven16crime/tot_crime16 *100
per_central16 = central16/tot_crime16 *100
per_dev16 = dev16/tot_crime16 *100
per_fhill16 = fhill16/tot_crime16 *100
per_harb16 = harb16/tot_crime16 *100
per_hbeck16 = hbeck16/tot_crime16 *100
per_hwood16 = hwood16/tot_crime16 *100
per_mission16 = mission16/tot_crime16 *100
per_nhwood16 = nhwood16/tot_crime16 *100
per_newton16 = newton16/tot_crime16 * 100
per_northeast16 = northeast16/tot_crime16 *100
per_pacific16 = pacific16/tot_crime16 *100
per_rampart16 = rampart16/tot_crime16 *100
per_seast16 = seast16/tot_crime16 *100
per_swest16 = swest16/tot_crime16 *100
per_topanga16 = topanga16/tot_crime16*100
per_vnuys16 = vnuys16/tot_crime16 *100
per_wla16 = wla16/tot_crime16 *100
per_wvalley16 = wvalley16/tot_crime16 *100
per_wilshire16 = wilshire16/tot_crime16 *100


rate16 = pd.DataFrame({"":['77th St', 'Central', 'Devonshire', 'Foothill', 'Harbor',
                          'Hollenbeck', 'Hollywood', 'Mission', 'North Hollywood',
                          'Newton', 'Northeast',' Pacific', 'Rampart', 'Southeast',
                          'Southwest', 'Topanga', 'Van Nuys', 'West LA', 'West Valley',
                          'Wilshire'],
                           "Percentage of Crime": [per_seventyseven16crime, per_central16, per_dev16, per_fhill16, 
                                                         per_harb16,per_hbeck16, per_hwood16, per_mission16, per_nhwood16,
                                                         per_newton16,per_northeast16,per_pacific16, per_rampart16,
                                                         per_seast16, per_swest16, per_topanga16,
                                                         per_vnuys16,
                                                         per_wla16, per_wvalley16, per_wilshire16]})

rate16["Percentage of Crime"]= rate16["Percentage of Crime"].map("{:.2f}%".format)
rate16= rate16.set_index('')
rate16
```
