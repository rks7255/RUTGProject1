```python
#Crating new csv for filttered data
crimedata_2018_1_df.to_csv("../Source/crimedata_2018_1_P.csv", index=False, header=True)

#Adding colunmn in new csv of crime data
crimedata_2018_1_df["Zip Codes"] = ""
crimedata_2018_1_df.head()
```

```python
#find out the xip codes after giving values of lat and lng

url = "https://maps.googleapis.com/maps/api/geocode/json?"
params={'key':gkey}
#set the counter for output checking purpose
counter = 1
print("Begining")
for index,row in crimedata_2018_1_df.iterrows():
    location = row["Location "].replace('(', '').replace(')', '')
    params["latlng"] = location
    params["result_type"] = "postal_code"
    response = requests.get(url,params=params).json()
    result = response['results']
    if(len(result)>0):
        if('address_components' in result[0].keys()):
            if(len(result[0]['address_components'])>0):
                if('long_name' in result[0]['address_components'][0].keys()):
                    #print("Updating zipcode for: {}".format(location))
                    print(f"Record : {counter}")
                    crimedata_2018_1_df.loc[index,"Zip Codes"]=result[0]['address_components'][0]['long_name']
                    #counter increment
                    counter+=1
                
    
                    
    
```
